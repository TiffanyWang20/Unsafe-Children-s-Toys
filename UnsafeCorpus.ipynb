
### Function to tokenize and match the tokens with the unsafed toy list

```python
# rewrite the counting function
def count_dpterm(row):    
    tf = WordCloud().process_text(row)
    # intersect based on corpus
    intersec = {x:tf[x] for x in unsafe_corpus if x in tf}
    # union to keep corpus structure
    freq = unsafe_corpus | intersec
    # freq is a dictionary of depress term and its frequency
    # add count of non-depression terms = number of tokens in a tweet - number of unique matched tokens with depress dict
    # freq['non_depress_count'] = len(tokens) - len(set(dtokens))
    return freq
```

```python
# count word freq by calling count_dpterm function for each member of tweets
word_count = list(map(count_dpterm, toy_detail))
```
