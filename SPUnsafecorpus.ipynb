```python
import pandas as pd
df = pd.read_csv('490UnsafeCorpus(70).csv', encoding='latin1')
    
```

```python
df.tail(10)
```

```python
# test with one row
text = df.Description[0]
print(text)
```

```python
# Filter out NaN (missing) values and convert everything to string
cleaned_descriptions = [str(description) for description in df.Description if pd.notna(description)]

# Join the cleaned descriptions into one large text
text = " ".join(cleaned_descriptions)

print("There are {} words in the combination of all descriptions.".format(len(text)))
```

```python
# Generate a word cloud image, max words = 200
wordcloud = WordCloud(background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```

```python
# select only unsafe toy products 
# change "safe" to create safe dictionary(Important)
df_unsafe = df[df['Label'] == "Unsafe"]
```

```python
# join all descriptions to one large text
text = " ".join(review.lower() for review in df_unsafe.Description)
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
# Assuming your string is stored in a variable named 'text'
cleaned_text = text.replace("", "")
```

```python
wordcloud = WordCloud(background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```

```python
# Create frequency dict
freq = wordcloud.process_text(text)
print(freq)
```

```python
# sort by value
sort_term = sorted(freq.items(), key=lambda x:x[1], reverse=True)
print(sort_term)
```

```python
# convert result to dict
sort_dict = dict(sort_term)
print(sort_dict)
```

```python
sort_dict_term = set(sort_dict.keys())
print(sort_dict_term)
```

```python
import pandas as pd
pd_sort = pd.DataFrame.from_dict(sort_dict, orient='index')
pd_sort.head(10)
```

```python
#safe&unsafe
pd_sort.to_csv('sort_dictSP.csv')
```
